In our initial analysis of the mobile game Cookie Cats, we investigated the 1 day retention proportion of players. Although there was strong evidence in favor of version gate 30 1 day retention may not be the best metric. After 1 day of play playrs may not have reached level 30 or lvel 40 inorder for there to be a comparison of their effects. In order to account for this we will run the same analysis but with 7 day retention values. 

```python
import pandas as pd

# Reading in the data
df = pd.read_csv("datasets/cookie_cats.csv")

%matplotlib inline

# Creating a list with bootstrapped means for each AB-group
boot_7d = []
for i in range(500):
    boot_mean = df.sample(frac=1, replace = True).groupby('version')['retention_1'].mean()
    boot_7d.append(boot_mean)
    
# Transforming the list to a DataFrame

boot_7d = pd.DataFrame(boot_7d)

# Adding a column with the % difference between the two AB-groups
boot_7d['diff'] = ((boot_7d['gate_30'] - boot_7d['gate_40'])/ boot_7d['gate_40']
)*100

# Ploting the bootstrap % difference
ax = boot_7d['diff'].plot.kde()
ax.set_xlabel("% difference in means")

# Calculating the probability that 7-day retention is greater when the gate is at level 30
prob = (boot_7d['diff']>0).mean()

# Pretty printing the probability

print( "{0:.0%}".format(prob))
```

Using bootstrap method to resample, we've constructed a KDE plot of the % difference between gate 30 and gate 40. From the KDE we can see that the majority of the difference are above 0 in favor of gate 30, in fact 98% of the differences were above 0.
